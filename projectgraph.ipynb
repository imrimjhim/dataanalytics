```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('imdb_top_1000.csv')
d=pd.DataFrame(df)
print(d)
print(d.head())
print(d.tail())
d.dropna(inplace=True)
print(d.info())
print(d.columns)
plt.bar(d['IMDB_Rating'].head(12),d['Genre'].head(12))
plt.title("top 12 movie rating")
plt.ylabel("Genre")
plt.xlabel("Movie rating")
plt.show()
plt.bar(d['Meta_score'].head(10),d['Series_Title'].head(10))
plt.title("Rating line plot")
plt.xlabel("MetaScore")
plt.ylabel("SeriesTitle")
plt.show()
```
